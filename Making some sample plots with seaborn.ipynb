```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
CPS_data_frame = pd.read_stata("CPS_data_even_years/CPS_data.dta",
                               convert_categoricals=False, 
                               convert_missing=False)
```

```python
%matplotlib inline
sns.set_context("talk",font_scale=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 5))
sns.distplot(CPS_data_frame['inctot'][
        (10000<CPS_data_frame['inctot']) & (CPS_data_frame['inctot']<500000)],
             color='r', ax=axes[0], bins=20, axlabel="Income")
axes[0].set_xticks([0,200000,400000,600000])
axes[0].ticklabel_format(axis='y', style='sci', scilimits=(-2,2))
sns.distplot(np.log(CPS_data_frame['inctot'][
        (10000<CPS_data_frame['inctot']) & (CPS_data_frame['inctot']<500000)]),
             color='b', ax=axes[1], bins=20, axlabel="Natural Log of Income")
axes[1].set_xticks([9,10,11,12,13])
sns.distplot(CPS_data_frame['age'],
             color='g', ax=axes[2], bins=20, axlabel="Age")
axes[2].set_xticks([0,20,40,60,80])
plt.tight_layout()
plt.savefig("example_histograms.pdf",bbox_inches='tight')
plt.show()
```
