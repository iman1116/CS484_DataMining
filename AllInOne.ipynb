```python
#PRE-PROCESSING 
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("talk",font_scale=2)
```

```python
np.random.seed(47)
```

```python
CPS_data_frame = pd.read_stata("CPS_data_even_years/CPS_data.dta",
                               convert_categoricals=False, 
                               convert_missing=False)
```

```python
CPS_data_frame = CPS_data_frame[(10000 < \
                                 CPS_data_frame['inctot']) & (CPS_data_frame['inctot'] < \
                                                              500000)] 
```

```python
CPS_data_frame = CPS_data_frame[CPS_data_frame['age'] >= 30] 
```

```python
CPS_data_frame['worker'] = (CPS_data_frame['labforce'] == 2).astype(int)
CPS_data_frame['female'] = (CPS_data_frame['sex'] == 2).astype(int)
CPS_data_frame['loginc'] = np.log(CPS_data_frame['inctot'])
```

```python
CPS_data_frame['college'] = (CPS_data_frame['educ99'] >= 15).astype(int)
```

```python
CPS_data_frame['masters'] = (CPS_data_frame['educ99'] == 16).astype(int)
```

```python
CPS_data_frame['high_school'] = (CPS_data_frame['educ99'] >= 10).astype(int)
```

```python
CPS_data_frame['professional_degree'] = (CPS_data_frame['educ99'] == 17).astype(int)
```

```python
CPS_data_frame['doctorate'] = (CPS_data_frame['educ99'] == 18).astype(int)
```

```python
CPS_data_frame['white'] = (CPS_data_frame['race'] == 100).astype(int)
```

```python
CPS_data_frame['not_hispanic'] = (CPS_data_frame['hispan'] == 0).astype(int)
```

```python
list_of_features = ['loginc', 'age', 'female', 'worker', 'white', 'not_hispanic',
                    'college', 'masters', 'high_school', 'professional_degree', 'doctorate']
```

```python
temp_df = CPS_data_frame[list_of_features]
```

```python
temp_df = temp_df[temp_df['age'] <= 34]
```

```python
temp_df = temp_df.dropna()
```

```python
X = temp_df.drop('loginc', 1)
y = temp_df['loginc']
```

```python
temp_df2 = CPS_data_frame[list_of_features]
temp_df2 = temp_df2[temp_df2['age'] <= 39]
temp_df2 = temp_df2[temp_df2['age'] >= 35]
temp_df2 = temp_df2.dropna()
```

```python
X_test = temp_df2.drop('loginc', 1).values
y_test = (temp_df2['loginc']).values
```

```python
from sklearn import tree
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold


regr_1 = tree.DecisionTreeRegressor(criterion = 'mse', 
                                    max_depth=4, 
                                    min_samples_split=25,
                                    min_samples_leaf=10)
regr_2 = tree.DecisionTreeRegressor(criterion = 'mse', 
                                    max_depth=6, 
                                    min_samples_split=25,
                                    min_samples_leaf=10)
regr_3 = tree.DecisionTreeRegressor(criterion = 'mse', 
                                    max_depth=6, 
                                    min_samples_split=25,
                                    min_samples_leaf=10)
regr_4 = tree.DecisionTreeRegressor(criterion = 'mse', 
                                    max_depth=8, 
                                    min_samples_split=75,
                                    min_samples_leaf=20)


kfold = KFold(n_splits=4, shuffle=True, random_state=47)

scores_for_regr_1 = cross_val_score(regr_1, X, y, cv=kfold)
scores_for_regr_2 = cross_val_score(regr_2, X, y, cv=kfold)
scores_for_regr_3 = cross_val_score(regr_3, X, y, cv=kfold)
scores_for_regr_4 = cross_val_score(regr_4, X, y, cv=kfold)
```

```python
regr_1.fit(X,y)
regr_2.fit(X,y)
regr_3.fit(X,y)
regr_4.fit(X,y)

scores_for_regr_1 = np.append(scores_for_regr_1, regr_1.score(X_test,y_test))
scores_for_regr_2 = np.append(scores_for_regr_2, regr_2.score(X_test,y_test))
scores_for_regr_3 = np.append(scores_for_regr_3, regr_3.score(X_test,y_test))
scores_for_regr_4 = np.append(scores_for_regr_4, regr_4.score(X_test,y_test))
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(3, 5))
sns.boxplot(data=[scores_for_regr_1, scores_for_regr_2, scores_for_regr_3, scores_for_regr_4])
plt.savefig("Regression_Tree_Boxplots.pdf",bbox_inches='tight')
plt.show()
```

```python
sns.set_context("talk",font_scale=1)
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(6, 5))
sns.boxplot(data=[scores_for_regr_1, scores_for_regr_2, scores_for_regr_3, scores_for_regr_4])
fig.get_axes()[0].set_xlabel("Model")
fig.get_axes()[0].set_xticklabels(["1", "2", "3", "4"])
fig.get_axes()[0].set_ylabel(r"$R^2$")
fig.get_axes()[0].set_title("Scores for the Regression Tree")
plt.savefig("Regression_Tree_Boxplots_small.pdf",bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
ESS_data_frame = pd.read_stata("ESS_Cumulative/ESS1-6e01_1_F1.dta",
                               convert_categoricals=False, 
                               convert_missing=False)
```

```python
ESS_data_frame = ESS_data_frame[ESS_data_frame['agea'] <= 53] 
ESS_data_frame = ESS_data_frame[ESS_data_frame['agea'] >= 30] 
```

```python
ESS_data_frame['female'] = (ESS_data_frame['gndr'] == 2).astype(int)

ESS_data_frame['born_in_country'] = (ESS_data_frame['brncntr'] == 1).astype(int)

ESS_data_frame['college'] = (ESS_data_frame['eisced'] == 6).astype(int)
ESS_data_frame['post_college'] = (ESS_data_frame['eisced'] == 7).astype(int)

ESS_data_frame['father_college'] = (ESS_data_frame['eiscedf'] == 6).astype(int)
ESS_data_frame['father_post_college'] = (ESS_data_frame['eiscedf'] == 7).astype(int)

ESS_data_frame['mother_college'] = (ESS_data_frame['eiscedm'] == 6).astype(int)
ESS_data_frame['mother_post_college'] = (ESS_data_frame['eiscedm'] == 7).astype(int)

ESS_data_frame['partner_college'] = (ESS_data_frame['eiscedp'] == 6).astype(int)
ESS_data_frame['partner_post_college'] = (ESS_data_frame['eiscedp'] == 7).astype(int)
```

```python
country_dummies = pd.get_dummies(ESS_data_frame['cntry'])
ESS_data_frame = pd.concat([ESS_data_frame, country_dummies], axis=1)
```

```python
ESS_features = ['hinctnta', 'agea', 'female', 'born_in_country', 'college', 'post_college',
                'father_college', 'father_post_college', 
                'mother_college', 'mother_post_college',
                'partner_college', 'partner_post_college'] + country_dummies.columns.tolist()
```

```python
ESS_temp_df = ESS_data_frame[ESS_features]
ESS_temp_df = ESS_temp_df.dropna()
ESS_temp_df = ESS_temp_df.T.groupby(level=0).first().T
```

```python
ESS_temp_df.shape
```

```python
from sklearn.model_selection import train_test_split

ESS_X_train, ESS_X_test, ESS_y_train, ESS_y_test = train_test_split(
    ESS_temp_df.drop('hinctnta', 1), ESS_temp_df['hinctnta'], random_state = 47)
```

Remember the definition of our choice of score:
$$ \frac{2 \times \text{true positive}}{2 \times \text{true positive} + \text{false negative} + \text{false positive}} $$

Here, for our multiclass classification problem, the "micro" averaging computes the total number of false positives, false negatives, and true positives over all classes, and then computes the f-score using these counts.

```python
class_tree_1 = tree.DecisionTreeClassifier(criterion ='entropy', 
                                           max_depth=7, 
                                           min_samples_split=25,
                                           min_samples_leaf=10)

class_tree_2 = tree.DecisionTreeClassifier(criterion ='gini', 
                                           max_depth=9, 
                                           min_samples_split=40,
                                           min_samples_leaf=15)

class_tree_3 = tree.DecisionTreeClassifier(criterion ='entropy', 
                                           max_depth=11, 
                                           min_samples_split=50,
                                           min_samples_leaf=15)

class_tree_4 = tree.DecisionTreeClassifier(criterion ='gini', 
                                           max_depth=11, 
                                           min_samples_split=100,
                                           min_samples_leaf=30)

kfold = KFold(n_splits=4, shuffle=True, random_state=47)

scores_for_class_tree_1 = cross_val_score(class_tree_1,
                                          ESS_X_train, ESS_y_train,
                                          cv=kfold, scoring='f1_micro')
scores_for_class_tree_2 = cross_val_score(class_tree_2,
                                          ESS_X_train, ESS_y_train,
                                          cv=kfold, scoring='f1_micro')
scores_for_class_tree_3 = cross_val_score(class_tree_3,
                                          ESS_X_train, ESS_y_train,
                                          cv=kfold, scoring='f1_micro')
scores_for_class_tree_4 = cross_val_score(class_tree_4,
                                          ESS_X_train, ESS_y_train,
                                          cv=kfold, scoring='f1_micro')
```

```python
class_tree_1.fit(ESS_X_train, ESS_y_train)
class_tree_2.fit(ESS_X_train, ESS_y_train)
class_tree_3.fit(ESS_X_train, ESS_y_train)
class_tree_4.fit(ESS_X_train, ESS_y_train)

scores_for_class_tree_1 = np.append(scores_for_class_tree_1,
                                    class_tree_1.score(ESS_X_test, ESS_y_test))
scores_for_class_tree_2 = np.append(scores_for_class_tree_2,
                                    class_tree_2.score(ESS_X_test, ESS_y_test))
scores_for_class_tree_3 = np.append(scores_for_class_tree_3,
                                    class_tree_3.score(ESS_X_test, ESS_y_test))
scores_for_class_tree_4 = np.append(scores_for_class_tree_4,
                                    class_tree_4.score(ESS_X_test, ESS_y_test))
```

```python
sns.set_context("talk",font_scale=1)
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(6, 5))
sns.boxplot(data=[scores_for_class_tree_1,
                  scores_for_class_tree_2,
                  scores_for_class_tree_3,
                  scores_for_class_tree_4])
fig.get_axes()[0].set_xlabel("Model")
fig.get_axes()[0].set_xticklabels(["1", "2", "3", "4"])
fig.get_axes()[0].set_ylabel("F1 score")
fig.get_axes()[0].set_title("Scores for the Decision Trees")
plt.savefig("Decision_Tree_Boxplots_small.pdf",bbox_inches='tight')
plt.tight_layout()
plt.show()
```
